# Markers

```python
map_bishkek = folium.Map(location=[42.872595, 74.591327],
                        tiles='Stamen Terrain',
                        zoom_start=12)
folium.Marker([42.872595, 74.591327], popup='Isanova Street').add_to(map_bishkek)
map_bishkek
```

```python
map_bishkek = folium.Map(location=[42.872595, 74.591327],
                        tiles='Stamen Terrain',
                        zoom_start=12)

#Simple marker
folium.Marker([42.872595, 74.591327],
             popup='Manas Avenue',
             icon=folium.Icon(color='green')
             ).add_to(map_bishkek)

#Circle marker
folium.CircleMarker([42.8760744,74.5857558],
                   radius=30,
                   popup='NBKR',
                   color='red',
                   ).add_to(map_bishkek)

#Interactive marker
map_bishkek.add_child(folium.ClickForMarker(popup='Bishkek is awesome'))
map_bishkek
```

# Interaction with the map

```python
map_bishkek = folium.Map(location=[42.872595, 74.591327],
                        tiles='Stamen Terrain',
                        zoom_start=12)

from folium import plugins
from folium.plugins import MeasureControl
map_bishkek.add_child(MeasureControl())

from folium.plugins import FloatImage
url = ('https://commons.wikimedia.org/wiki/File:Bishkek.jpg')
FloatImage(url, bottom=5, left=85).add_to(map_bishkek)

map_bishkek
```

# Heatmaps

```python
df = pd.read_csv('test.csv')
df .head()
df['pickup_latitude'] = df['pickup_latitude'].astype(float)
df['pickup_longitude'] = df['pickup_longitude'].astype(float)
df.head()
```

```python
# Heatmap

from folium import plugins
from folium.plugins import HeatMap

map_bishkek = folium.Map(location=[40.763805, -73.973320],
                        tiles='Stamen Terrain',
                        zoom_start=11)

heat_data = [[row['pickup_latitude'], row['pickup_longitude']] for index, row in df.iterrows()]

HeatMap(heat_data).add_to(map_bishkek)
map_bishkek
```
